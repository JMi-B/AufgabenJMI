```python
import pandas as pd
import matplotlib.pyplot as plt

# Pfad zur CSV-Datei
pfad = "../data-prozess-20250714/Daten und Prozessanalyse/Woche 2/Tag 8/csv/hardwarebestellung_prozesse.csv"  # ‚Üê ersetze durch deinen Dateinamen

# Einlesen der Datei
df = pd.read_csv(pfad)

# Ausgabe der ersten 5 Zeilen zur Kontrolle

print(df.head())    # Erste Zeilen anzeigen
print(df.tail())     # Letzte Zeilen anzeigen
print(df.columns)      # Spaltennamen pr√ºfen
print(df.shape)      # Dimension (Zeilen, Spalten)
print(df.dtypes)       # Datentypen pr√ºfen

print("isnull")
print(df.isnull().sum() )               # Anzahl fehlender Werte pro Spalte

print("isnull zeile")
print(df[df.isnull().any(axis=1)])


print("Abteilungen")
print(df["Abteilung"].value_counts())     # H√§ufigkeit pro Kategorie
print("R√ºckfragen")
print(df["Anzahl_Rueckfragen"].value_counts())     # H√§ufigkeit pro Kategorie


print("Anfragedatum")
print(pd.to_datetime(df["Anfragedatum"], errors="coerce").isnull().sum())

print("Uebergabedatum")
print(pd.to_datetime(df["Uebergabedatum"], errors="coerce").isnull().sum())

df["dauer"] = (pd.to_datetime(df["Uebergabedatum"]) - pd.to_datetime(df["Anfragedatum"])).dt.days

# Statistische Kennzahlen anzeigen
print("üìä Statistische Kennzahlen:")
print(df["dauer"].describe())

# # Boxplot anzeigen
# plt.figure(figsize=(5, 4))
# df["dauer"].plot.box()
# plt.title("Boxplot der Durchlaufdauer")
# plt.ylabel("Tage")
# plt.grid(True)
# plt.tight_layout()
# plt.show()

print(df.duplicated().sum() )                  # Anzahl doppelter Zeilen
print(df[df.duplicated(subset="Prozess_ID")])

# print(df["dauer"].describe())          # Statistische Kennzahlen
# print(df["dauer"].plot.box())   

# check auf doppelte spaltennamen (keine Unterscheidung gro√ü / kleinschreibung)
print(df[df["dauer"]<0].sum())
df["ABTEILUNG"] = df["Abteilung"]

df["prio"] = df["Prioritaet"]
df["Prio"] = df["Prioritaet"]

a = df.columns.str.upper()
print(a)
b = a[a.duplicated()].unique().to_list()
print(b)

```
